```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('unemployment analysis.csv' , sep=",")
df1
```

```python
df2 = pd.read_csv('continents2.csv' , sep=",")
df2
```

```python
df2temp=df2.loc[ : , [ 'alpha-3', 'region','sub-region', 'intermediate-region'] ]
```

```python
df2temp.columns= ['Country Code', 'region', 'sub-region', 'intermediate-region']
```

```python
df2temp
```

```python
dfnew = pd.DataFrame()
for y in df1.columns[ 2: ]:
    dftemp = df1.loc[ : , ["Country Name","Country Code",y] ]
    dftemp["Year"] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Value'] ]
    
    dfnew = pd.concat([dfnew, dftemp])

dfnew
```

```python
df_new=dfnew.merge(df2temp,how='left', on='Country Code')
df_new.columns= ['Country Name', 'Country Code', 'Year', 'Value', 'Continent', 'Region','intermediate-region']
df_new
```
