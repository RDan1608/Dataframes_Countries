```python
import pandas as pd
import numpy as np
```

# Dataframe Unemployment Rate

```python
df1 = pd.read_csv('unemployment analysis.csv' , sep=",")
df1
```

```python
dfnew = pd.DataFrame()
for y in df1.columns[ 2: ]:
    dftemp = df1.loc[ : , ["Country Name","Country Code",y] ]
    dftemp["Year"] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Value'] ]
    
    dfnew = pd.concat([dfnew, dftemp])

dfnew
```

# Dataframe Currency and Income Information

```python
df2 = pd.read_csv('Country.csv' , sep=",")
df2
```

## Obtenemos las columnas que necesitamos y su codigo de pais para hacer el Merge

```python
df2temp=df2.loc[ : , ['CountryCode','CurrencyUnit', 'Region', 'IncomeGroup'] ]
```

```python
df2temp.columns= ['Country Code','CurrencyUnit', 'SubRegion', 'IncomeGroup']
```

```python
df2temp
```

# Dataframe Continents

```python
df3 = pd.read_csv('continents2.csv' , sep=",")
df3
```

## Obtenemos las columnas que necesitamos y su codigo de pais para hacer el Merge

```python
df3temp=df3.loc[ : , [ 'alpha-3', 'region'] ]
```

```python
df3temp.columns= ['Country Code', 'Continent']
```

```python
df3temp
```

# Nuevo Dataframe (Merge df1 y df2) 

## Se realizara un merge para unir los dataframes de unemployment rate y currency and income information

```python
df_new=dfnew.merge(df2temp,how='left', on='Country Code')
df_new.columns= ['Country Name', 'Country Code', 'Year', 'Value', 'CurrencyUnit', 'SubRegion', 'IncomeGroup']
df_new
```

# Dataframe final (Merge dfnew y df3)

## Se realizara un merge entre el dataframe anterior(dfnew) y el dataframe continents para añadir la informacion de los continetes

```python
df_1new=df_new.merge(df3temp,how='left', on='Country Code')
df_1new.loc[ : , ['Country Name', 'Country Code','Continent','SubRegion', 'Year', 'Value' ,'CurrencyUnit', 'IncomeGroup']]
```
